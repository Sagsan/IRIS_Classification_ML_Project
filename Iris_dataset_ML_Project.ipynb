```python
# Suppress warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Set style and plot
sns.set_style('whitegrid')
sns.pairplot(data=df, hue='Species')
```

```python
# Create the initial scatter plot for 'setosa'
pet_len_wid = df[df.Species == 'Iris-setosa'].plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm',
                                                    color='blue', label='setosa')

# Plot 'versicolor' on the same axes
df[df.Species == 'Iris-versicolor'].plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm',
                                         color='red', label='versicolor', ax=pet_len_wid)

# Plot 'virginica' on the same axes
df[df.Species == 'Iris-virginica'].plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm', 
                                        color='green', label='virginica', ax=pet_len_wid)

# Set labels and title
pet_len_wid.set_xlabel('Sepal Length (cm)')
pet_len_wid.set_ylabel('Sepal Width (cm)')
pet_len_wid.set_title('Sepal Length vs Sepal Width for Different Iris Species')

# Set the figure size
pet_len_wid.figure.set_size_inches(10, 5)

# Show the plot
plt.show()
```

```python
# Create the initial scatter plot for 'setosa'
plt.figure(figsize=(10, 5))
plt.scatter(setosa['PetalLengthCm'], setosa['PetalWidthCm'], color='blue', label='setosa')

# Plot 'versicolor' on the same axes
plt.scatter(versicolor['PetalLengthCm'], versicolor['PetalWidthCm'], color='red', label='versicolor')

# Plot 'virginica' on the same axes
plt.scatter(virginica['PetalLengthCm'], virginica['PetalWidthCm'], color='green', label='virginica')

# Set labels and title
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width for Different Iris Species')
plt.legend()

# Show the plot
plt.show()
```
